```python
df = pd.DataFrame(sales_data)

df
```

## Dataset Description

The dataset contains simple product sales information.

Each row represents one product.

Columns:

- Product: product name
- Category: product category
- Region: sales region
- Price: selling price per unit
- Cost: cost per unit
- Quantity: number of units sold

```python
df.head()
```

```python
df.shape
```

```python
df.info
```

```python
df.describe
```

```python
df["Revenue"] = df["Price"] * df["Quantity"]
df["Total_cost"] = df["Cost"] * df["Quantity"]
df["Profit"] = df["Revenue"] - df["Total_cost"]

df
```

```python
df["Profit_Margin"] = df["Profit"] / df["Revenue"] * 100

df
```

```python
df["Profit_Margin"] = df["Profit_Margin"].round(2)

df
```

```python
total_revenue = df["Revenue"].sum()
total_cost = df["Total_cost"].sum()
total_profit = df["Profit"].sum()

print(f"Total Revenue: ${total_revenue}")
print(f"Total Cost: ${total_cost}")
print(f"Total Profit: ${total_profit}")
```

```python
top_revenue_product = df.sort_values("Revenue", ascending=False)

top_revenue_product
```

```python
best_product = top_revenue_product.iloc[0]

print("Best Product by Revenue")
print(f"Product: {best_product['Product']}")
print(f"Revenue: ${best_product['Revenue']}")
```

```python
top_profit_product = df.sort_values("Profit", ascending=False)

top_profit_product
```

```python
best_profit_product = top_profit_product.iloc[0]

print("Best Product by Profit")
print(f"Product: {best_profit_product['Product']}")
print(f"Profit: ${best_profit_product['Profit']}")
```

```python
category_summary = df.groupby("Category")[["Revenue", "Profit", "Quantity"]].sum()

category_summary
```

```python
category_summary.sort_values("Revenue", ascending=False)
```

```python
region_summary = df.groupby("Region")[["Revenue","Profit","Quantity"]].sum()

region_summary
```

```python
region_summary.sort_values("Profit", ascending=False)
```

```python
def classify_performance(profit):
    if profit >= 1000:
        return "High"
    elif profit >= 500:
        return "Medium"
    else:
        return "Low"

df["Performance"] = df["Profit"].apply(classify_performance)

df
```

```python
high_performance_products = df[df["Performance"] == "High"]

high_performance_products
```

```python
low_performance_products = df[df["Performance"] ==  "Low"]

low_performance_products
```

```python
df.plot(
    x="Product",
    y="Revenue",
    kind="bar",
    title="Revenue by Product"
)

plt.xlabel("Product")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
category_summary["Profit"].plot(
    kind="bar",
    figsize=(8, 5),
    title="Profit by Category"
)

plt.xlabel("Category")
plt.ylabel("Profit")
plt.xticks(rotation=0)
plt.show()
```

```python
df.to_csv("processed_sales_data_csv", index=False)
print("Processed sales data saved successfully.")
```
