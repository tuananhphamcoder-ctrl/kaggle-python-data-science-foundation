# Mini Sales Analyzer with Python

## Project Overview

This is my first Python project for data science learning.

The goal of this project is to analyze simple product sales data using Python basics such as:

- Variables
- Lists
- Dictionaries
- Loops
- Functions
- Conditional statements

This project does not use Pandas yet. It focuses on Python fundamentals before moving to data analysis and machine learning.

```python
products = [
    {
        "name": "Laptop",
        "price": 1200,
        "quantity": 5
    },
    {
        "name": "Mouse",
        "price": 25,
        "quantity": 50
    },
    {
        "name": "Keyboard",
        "price": 75,
        "quantity": 20
    },
    {
        "name": "Monitor",
        "price": 300,
        "quantity": 10
    },
    {
        "name": "Headphones",
        "price": 100,
        "quantity": 15
    }
]

products
```

## Dataset Description

The dataset contains product sales information.

Each product has:

- Product name
- Price
- Quantity sold

Revenue is calculated using the formula:

`Revenue = Price × Quantity`

```python
for product in products:
    revenue = product["price"] * product["quantity"]
    product["revenue"] = revenue

products
```

```python
print("Sales Revenue Report")
print("-" * 30)

for product in products:
    print(f"{product['name']}: ${product['revenue']}")
```

```python
total_revenue = 0

for product in products:
    total_revenue += product["revenue"]

print(f"Total Revenue: ${total_revenue}")
```

```python
best_product = None
highest_revenue = 0

for product in products:
    if product["revenue"] > highest_revenue:
        highest_revenue = product["revenue"]
        best_product = product["name"]

print(f"Best Product: {best_product}")
print(f"Highest Revenue: ${highest_revenue}")
```

```python
print("Final Sales Analysis Report")
print("=" *40)

for product in products:
    print(f"Product: {product['name']}")
    print(f"Price: ${product['price']}")
    print(f"Quantity Sold: {product['quantity']}")
    print(f"Revenue: ${product['revenue']}")
    print(f"Performance: {product['performance']}")
    print("_" *40)

print(f"Total Revenue: ${total_revenue}")
print(f"Average Revenue: ${average_revenue:.2f}")
print(f"Best product: {best_product}")
```
